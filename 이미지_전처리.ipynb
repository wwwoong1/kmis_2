```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import io
folder_path = "/content/drive/MyDrive/크롤링 이미지/풍경"
file_list = os.listdir(folder_path)
for i in range(0,4925):
    # Image Load
    image = io.imread(img_list3[i])
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    save_file = "/content/drive/MyDrive/풍경모델/풍경전처리/풍경실사/" + file_list[i]
    # Sharpening function
    def sharpening(image, strength):
        b = (1 - strength) / 8
        sharpening_kernel = np.array([[b, b, b],
                                    [b, strength, b],
                                    [b, b, b]])
        output = cv2.filter2D(image, -1, sharpening_kernel)
        return output

    # Result
    output1 = sharpening(gray_image, strength=17)

    # view
    plt.subplot(1, 2, 1)
    plt.imshow(gray_image, cmap='gray')
    plt.title('Original')
    plt.axis('off')
    plt.subplot(1, 2, 2)
    plt.imshow(output1, cmap='gray')
    plt.title('strength=17')
    plt.axis('off')
    cv2.imwrite(save_file, output1)
```

# Canny

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
folder_path = "/content/drive/MyDrive/크롤링 이미지/고양이"
file_list = os.listdir(folder_path)
for i in range(0,2204):
    image_gray = cv2.imread(img_list1[i], cv2.IMREAD_GRAYSCALE) # 흑백 이미지로 로드
    median_intensity = np.median(image_gray) # 픽셀 강도의 중간값을 계산
    save_file = "/content/drive/MyDrive/canny/real/cat/" + file_list[i]

    # 중간 픽셀 강도에서 위아래 1 표준 편차 떨어진 값을 임계값으로 지정합니다.
    lower_threshold = int(max(0, (1.0 - 0.33) * median_intensity))
    upper_threshold = int(min(255, (1.0 + 0.33) * median_intensity))

    # 캐니 경계선 감지기를 적용합니다.
    image_canny = cv2.Canny(image_gray, lower_threshold, upper_threshold)

    cv2.imwrite(save_file, image_canny)
```
